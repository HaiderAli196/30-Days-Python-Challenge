### Time Complexity First Method


```python
import time

start=time.time()


for i in range(1,101):
    print(i)

print(time.time() -start)    
```

### Now Changing the algorithm with same logic

```python
import time

start=time.time()
i=1
while i<101:
    i+=1
    print(i)
print(time.time()-start)    
```

```python
# Problem 1

L=[1,2,3,4,5]
sum=0
for i in L:
    sum=sum+i
print(sum)

product=1
for i in L:
    product=product*i
print(product)    
```

### Time Complexity of the above program 
### is given as O(n) +O(n) =O(2n)=O(n)
### There is a linear relationship

```python
L=[1,2,3,4,5]
for i in L:
    for j in L:
        print(('{},{}'.format(i,j)))

#In this example we have two loops but one loop is nested inside the other
#one so in this case the time complexity is given by O(n*n)=O(n^2) 
# This is a quadratic relation             
```

###Linear Search :
As it is obvious from its name it is a linear relation so the time complexity of this algorithm will be Order of n as  O(n)###
